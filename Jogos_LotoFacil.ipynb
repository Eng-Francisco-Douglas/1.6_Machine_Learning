```python
# Importar as bibliotecas necessárias
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier

# Criar um conjunto de dados sintético com os resultados de 20 jogos anteriores
# Cada jogo é representado por um vetor binário de 25 posições, onde 1 indica que o número foi sorteado e 0 que não foi
# Os números sorteados são escolhidos aleatoriamente entre 1 e 25
np.random.seed(42) # Para garantir a reprodutibilidade
X = np.zeros((20, 25)) # Matriz de 20 linhas e 25 colunas
for i in range(20):
    X[i, np.random.choice(25, 15, replace=False)] = 1 # Sortear 15 números para cada jogo

# Visualizar os dados em um dataframe
df = pd.DataFrame(X, columns=range(1, 26))
df.index.name = "Jogo"
print(df)

# Dividir os dados em treino e teste
X_train = X[:15] # 15 jogos para treino
X_test = X[15:] # 5 jogos para teste

# Criar e treinar um modelo de classificação usando florestas aleatórias
model = RandomForestClassifier(n_estimators=100)
model.fit(X_train, X_train)

# Fazer previsões para os dados de teste e de treino
y_pred_test = model.predict(X_test)
y_pred_train = model.predict(X_train)

# Avaliar o desempenho do modelo usando a acurácia (porcentagem de acertos)
acc_test = np.mean(y_pred_test == X_test)
acc_train = np.mean(y_pred_train == X_train)
print(f"Acurácia para os dados de teste: {acc_test:.2f}")
print(f"Acurácia para os dados de treino: {acc_train:.2f}")

# Fazer uma previsão para um novo jogo
y_pred_new = model.predict(np.zeros((1, 25))) # Prever a partir de um vetor vazio
y_pred_new = np.where(y_pred_new == 1)[1] + 1 # Obter os índices dos números sorteados e somar 1
print(f"Previsão para um novo jogo: {y_pred_new}")
```

```python
# Importar bibliotecas
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

# Baixar os dados da LotoFacil da internet
url = "http://www1.caixa.gov.br/loterias/_arquivos/loterias/D_lotfac.zip"
df = pd.read_csv(url, sep=";", encoding="latin1", skiprows=1)

# Selecionar as colunas com os números sorteados
cols = ["Bola1", "Bola2", "Bola3", "Bola4", "Bola5", "Bola6", "Bola7", "Bola8", "Bola9", "Bola10", "Bola11", "Bola12", "Bola13", "Bola14", "Bola15"]

# Transformar os dados em uma matriz binária, onde cada linha representa um sorteio e cada coluna representa um número de 1 a 25
# Se o número foi sorteado, o valor é 1, caso contrário é 0
X = np.zeros((len(df), 25))
for i in range(len(df)):
    for j in range(15):
        X[i][df[cols[j]][i]-1] = 1

# Dividir os dados em conjuntos de treino e teste
X_train, X_test = train_test_split(X, test_size=0.2, random_state=42)

# Criar um modelo de regressão logística para prever a probabilidade de cada número ser sorteado
model = LogisticRegression(max_iter=1000)
model.fit(X_train, X_train)

# Gerar 10 jogos futuros da LotoFacil, escolhendo os 15 números com maior probabilidade em cada jogo
predictions = model.predict_proba(X_test[:10])
games = []
for i in range(10):
    game = []
    probs = predictions[i]
    indices = np.argsort(probs)[-15:]
    for j in indices:
        game.append(j+1)
    games.append(game)

# Mostrar os jogos gerados
print("Os jogos gerados são:")
for i in range(10):
    print(f"Jogo {i+1}: {games[i]}")
```

```python

# Importar as bibliotecas necessárias
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Definir os parâmetros do modelo
NUMEROS = 25 # Número de números possíveis na Lotofacil
SEQUENCIA = 15 # Número de números sorteados em cada jogo
ENTRADA = 100 # Número de jogos anteriores usados como entrada
SAIDA = 10 # Número de jogos futuros a serem previstos
EPOCAS = 50 # Número de épocas de treinamento
UNIDADES = 64 # Número de unidades na camada LSTM

# Carregar os dados históricos da Lotofacil
df = pd.read_excel("Pasta1.xlsx") # Arquivo CSV com os dados históricos
df = df.iloc[:, 2:] # Remover as colunas de data e concurso
df = df.apply(lambda x: x - 1) # Subtrair 1 de cada número para ficar entre 0 e 24

# Converter os dados em arrays numpy
dados = df.values # Array bidimensional com os números sorteados em cada jogo
X = [] # Array tridimensional com as entradas do modelo (jogos anteriores)
y = [] # Array bidimensional com as saídas do modelo (jogos futuros)

# Criar as entradas e saídas do modelo a partir dos dados
for i in range(len(dados) - ENTRADA - SAIDA + 1):
  X.append(dados[i:i+ENTRADA]) # Pegar ENTRADA jogos consecutivos como entrada
  y.append(dados[i+ENTRADA:i+ENTRADA+SAIDA]) # Pegar SAIDA jogos seguintes como saída

# Converter as entradas e saídas em arrays numpy
X = np.array(X) # Array com shape (len(dados) - ENTRADA - SAIDA + 1, ENTRADA, SEQUENCIA)
y = np.array(y) # Array com shape (len(dados) - ENTRADA - SAIDA + 1, SAIDA, SEQUENCIA)

# Transformar os números em vetores one-hot encoding
X = tf.keras.utils.to_categorical(X, num_classes=NUMEROS) # Array com shape (len(dados) - ENTRADA - SAIDA + 1, ENTRADA, SEQUENCIA, NUMEROS)
y = tf.keras.utils.to_categorical(y, num_classes=NUMEROS) # Array com shape (len(dados) - ENTRADA - SAIDA + 1, SAIDA, SEQUENCIA, NUMEROS)

# Construir o modelo de rede neural recorrente (RNN)
modelo = Sequential()
modelo.add(LSTM(UNIDADES, input_shape=(ENTRADA, SEQUENCIA, NUMEROS))) # Camada LSTM com UNIDADES unidades e entrada com shape (ENTRADA, SEQUENCIA, NUMEROS)
modelo.add(Dense(SAIDA * SEQUENCIA * NUMEROS, activation="softmax")) # Camada Dense com SAIDA * SEQUENCIA * NUMEROS neurônios e função de ativação softmax
modelo.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"]) # Compilar o modelo com função de perda entropia cruzada categórica, otimizador adam e métrica acurácia

# Treinar o modelo com os dados de entrada e saída
modelo.fit(X, y.reshape(-1, SAIDA * SEQUENCIA * NUMEROS), epochs=EPOCAS) # Treinar o modelo por EPOCAS épocas e redimensionar a saída para um vetor unidimensional

# Prever os próximos 10 jogos da Lotofacil a partir dos últimos 100 jogos
X_teste = X[-1] # Pegar a última entrada do array X
y_pred = modelo.predict(X_teste.reshape(1, ENTRADA, SEQUENCIA, NUMEROS)) # Prever a saída a partir da entrada redimensionada para um vetor unidimensional
y_pred = y_pred.reshape(SAIDA, SEQUENCIA, NUMEROS) # Redimensionar a saída para um array tridimensional
y_pred = np.argmax(y_pred, axis=2) # Pegar o índice do maior valor em cada vetor one-hot encoding
y_pred = y_pred + 1 # Somar 1 a cada número para ficar entre 1 e 25

# Mostrar os próximos 10 jogos previstos pelo modelo
print("Os próximos 10 jogos previstos pelo modelo são:")
for i in range(SAIDA):
  print(f"Jogo {i+1}: {sorted(y_pred[i])}") # Mostrar cada jogo ordenado crescentemente

```
