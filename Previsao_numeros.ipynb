```python
# Importar as bibliotecas necessárias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Criar um conjunto de dados sintético com uma tendência linear
np.random.seed(42) # Para garantir a reprodutibilidade
X = np.arange(1, 101) # Variável independente
y = 2 * X + 5 + np.random.randn(100) * 10 # Variável dependente com ruído

# Visualizar os dados em um gráfico de dispersão
plt.scatter(X, y)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Dados sintéticos com tendência linear")
plt.show()

# Dividir os dados em treino e teste
X_train = X[:80].reshape(-1, 1) # 80% dos dados para treino
y_train = y[:80]
X_test = X[80:].reshape(-1, 1) # 20% dos dados para teste
y_test = y[80:]

# Criar e treinar um modelo de regressão linear
model = LinearRegression()
model.fit(X_train, y_train)

# Fazer previsões para os dados de teste e de treino
y_pred_test = model.predict(X_test)
y_pred_train = model.predict(X_train)

# Avaliar o desempenho do modelo usando o coeficiente de determinação (R²)
r2_test = model.score(X_test, y_test)
r2_train = model.score(X_train, y_train)
print(f"R² para os dados de teste: {r2_test:.2f}")
print(f"R² para os dados de treino: {r2_train:.2f}")

# Visualizar as previsões e os dados reais em um gráfico de dispersão
plt.scatter(X_test, y_test, label="Dados reais")
plt.scatter(X_test, y_pred_test, label="Previsões")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Previsões e dados reais para os dados de teste")
plt.legend()
plt.show()

# Fazer uma previsão para um valor futuro de X
X_future = 120 # Valor futuro de X
y_future = model.predict([[X_future]]) # Previsão para o valor futuro de X
print(f"Previsão para X = {X_future}: y = {y_future[0]:.2f}")
```
